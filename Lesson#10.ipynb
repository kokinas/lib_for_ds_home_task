```python
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv('train (1).csv')
df
```

```python
df.describe()
```

Имеем не корректные даты постройки

```python
df.loc[df['HouseYear'] > 2020, 'HouseYear'] = df['HouseYear'].median()
```

```python
df.info()
```

Healthcare_1 и LifeSquare имеют много пропусков, удалим эти столбцы

```python
target = df['Price']
df.drop(['Id', 'Healthcare_1', 'LifeSquare', 'Price'], axis=1, inplace=True)
```

```python
df = pd.get_dummies(df)
```

Модель линейной регрессии дало результаты по метрике r2 не более 0,4. Будем использовать случайный лес

```python
rgr = GridSearchCV(
    estimator= RandomForestRegressor(),
    scoring='r2',
    cv=5,
    param_grid={
        'n_estimators' : [100, 150, 200],
        'max_depth' : [9, 10, 11],
        'max_features' : np.arange(4,10)
    }
)
```

```python
%%time
X_train, X_test, y_train, y_test = train_test_split(df, target, test_size=0.3, random_state=27)
rgr.fit(X_train, y_train)
r2_score(y_test, rgr.predict(X_test)), r2_score(y_train, rgr.predict(X_train))
```

```python
cv_results = pd.DataFrame(rgr.cv_results_)
cv_results.sort_values('rank_test_score')
```

Тестовая выборка дала приемлимый резульата r2 = 0.74.

```python
X_finish = pd.read_csv('test.csv')
Id = X_finish['Id']
X_finish.drop(['Id', 'Healthcare_1', 'LifeSquare'], axis=1, inplace=True)
X_finish = pd.get_dummies(X_finish)
X_finish
```

```python
finish_pred = rgr.predict(X_finish)
```

```python
predict_prices = pd.DataFrame({'Id' : Id, 'Price': finish_pred})
predict_prices
```

```python
predict_prices.to_csv('finish_predict_prices', index=False)
```
