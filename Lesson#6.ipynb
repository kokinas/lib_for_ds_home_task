```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
data_boston = load_boston()
type(data_boston)
```

```python
print(data_boston.keys())
```

```python
print(len(data_boston.target))
print(data_boston.target)
target = data_boston.target
```

```python
print(data_boston.DESCR)
```

```python
data = data_boston.data
data 
```

```python
feature_names = load_boston().feature_names
feature_names
```

```python
df = pd.DataFrame(data, columns=feature_names)
df
```

```python
train_x, test_x, train_y, test_y = train_test_split(df, target, test_size=0.3, random_state=42)
len(train_y), len(test_y), len(test_y)/(len(test_y) + len(train_y))
```

```python
lr = LinearRegression()
```

```python
lr.fit(train_x, train_y)
```

```python
y_pred = lr.predict(test_x)
check_test = pd.DataFrame({'test_y': test_y, 'pred_y': y_pred})
check_test
```

```python
r2_score(check_test['test_y'], check_test['pred_y'])
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
pd.options.display.max_columns = 100
```

```python
df = pd.read_csv('../creditcard.csv')
```

```python
df.head(10)
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Class'].value_counts()
```

```python
X = df[df.columns.difference(['Class'])]
X
```

```python
y = df['Class']
y, type(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
parameters = [{'n_estimators': [10, 15],
'max_features': np.arange(3, 5),
'max_depth': np.arange(4, 7)}]
```

```python
model = GridSearchCV(estimator=RandomForestClassifier(random_state=100), param_grid=parameters, scoring='roc_auc',cv=3)
```

```python
%%time
model.fit(X_train, y_train)
```

```python
cv_results = pd.DataFrame(model.cv_results_)
cv_results
```

```python
model.best_params_
```

```python
model.predict_proba(X_test)
```

```python
y_pred_proba = model.predict_proba(X_test)[:,1]
y_pred_proba
```

```python
from sklearn.metrics import roc_auc_score
```

```python
len(y_test), len(y_pred_proba)
```

```python
roc_auc_score(y_test, y_pred_proba)
```

```python
roc_auc_score(y_test, y_pred_proba)
```
